```python
#Amostras dependentes - para a mesma serie temporal
#di=Xi-Yi
#The one-sided test has the null hypothesis that the median is positive against the alternative
#that it is negative (alternative == 'less'),
H0= 'Xi é maior ou igual a Yi'
H1= ' Xi é menor que Yi'

a = wilcoxon(Xi, Yi, zero_method='wilcox', correction=False, alternative='less', mode='auto')
a.pvalue
if(a.pvalue < alfa):
    print('Rejeita H0 e aceita H1:', H1)
else:
    print('Rejeita H1 e aceita H0:', H0)
a.pvalue                
                
```
